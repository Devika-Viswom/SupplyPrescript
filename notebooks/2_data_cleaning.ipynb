```python
import pandas as pd
```

```python
data_path = '../data'
```

```python
df=pd.read_csv(data_path + '/raw/raw1.csv')
```

```python
df_model=df.copy()
```

```python
df_model.drop_duplicates(inplace=True)
df_model["Date"] = pd.to_datetime(df_model["Date"])
```

```python
df_model.info()
```

```python
df_model["Year"] = df_model["Date"].dt.year
df_model["Month"] = df_model["Date"].dt.month
df_model["Quarter"] = df_model["Date"].dt.quarter
```

```python
df_model["Route"] = (
    df_model["Origin_Port"]
    + "_to_" +
    df_model["Destination_Port"]
)
```

```python
df_model["Reliability_Level"] = pd.cut(
    df_model["Carrier_Reliability_Score"],
    bins=[0,0.65,0.85,1],
    labels=["Low","Medium","High"]
)
```

```python
df_model["Risk_Level"] = pd.cut(
    df_model["Geopolitical_Risk_Score"],
    bins=[0,3,7,10],
    labels=["Low","Medium","High"]
)
```

```python
df_model.to_pickle(data_path + "/processed/df_model.pkl")
```
