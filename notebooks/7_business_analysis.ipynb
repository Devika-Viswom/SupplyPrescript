```python
import pandas as pd
```

```python
data_path = "../data"
```

```python
df_model=pd.read_pickle(data_path + '/processed/df_model.pkl')
```

```python
pd.cut(
    df_model["Geopolitical_Risk_Score"],
    bins=[0,3,7,10],
    labels=["Low","Medium","High"]
).to_frame("Risk_Bucket").join(
    df_model["Disruption_Occurred"]
).groupby("Risk_Bucket")[
    "Disruption_Occurred"
].mean()*100
```

```python
pd.cut(
    df_model["Carrier_Reliability_Score"],
    bins=[0,0.65,0.85,1],
    labels=["Low","Medium","High"]
).to_frame("Reliability_Bucket").join(
    df_model["Disruption_Occurred"]
).groupby("Reliability_Bucket")[
    "Disruption_Occurred"
].mean()*100
```

```python
df_model.groupby(
    "Transport_Mode"
)["Lead_Time_Days"].mean()
```

```python
df_model.groupby(
    "Weather_Condition"
)["Lead_Time_Days"].mean()
```

```python
df_model.groupby("Transport_Mode")[
    ["Distance_km","Weight_MT","Lead_Time_Days"]
].mean()
```
